```python
def getUniformPoints(num, dimension=2):
    i = 1
    all_fo = open("../../../datasets/uniform_" + str(num) + "_.csv", "w")
    for i in range(num):
        node_string = ''
        for dim in range(dimension):
            x = random.uniform(0, 1)
            node_string = node_string + str(x) + "," 
        node_string = node_string + str(i) + "\n"
        all_fo.write(node_string)
    all_fo.close()

def getNormalPoints(num, dimension=2):
    seeds = []
    for i in range(dimension):
        seeds.append(tf.truncated_normal([num,1], mean=0.5, stddev=0.25, dtype=tf.float32))
    with tf.Session() as sees:
        items = []
        for i in range(dimension):
            items.append(sees.run(seeds[i]))
        with open("../../../datasets/normal_" + str(num) + "_.csv", "w") as fo:
            for i in range(num):
                node_string = ''
                for item in items:
                    node_string = node_string + str(item[i][0]) + "," 
                node_string = node_string + str(i) + "\n"
                fo.write(node_string)

def getSkewedPoints(num, a, dimension=2):
    seeds = []
    for i in range(dimension):
        seeds.append(tf.truncated_normal([num,1], mean=0.5, stddev=0.25, dtype=tf.float32))
    with tf.Session() as sees:
        items = []
        for i in range(dimension):
            items.append(sees.run(seeds[i]))
        with open("../../../datasets/skewed_" + str(num) + "_skewness_" + str(a) + "_.csv", "w") as fo:
            for i in range(num):
                node_string = ''
                for item in items:
                    if item is items[-1]:
                        node_string = node_string + str(item[i][0]**a) + "," 
                    else:
                        node_string = node_string + str(item[i][0]) + "," 
                        
                node_string = node_string + str(i) + "\n"
                fo.write(node_string)                


if __name__ == '__main__':
    # getUniformRandomPoints(80000, 10)
    # get_tf_normalrandomPoints(8000000, 10)
    size = 2000000
#     getUniformPoints(size, 3)
    getNormalPoints(size, 2)
#     getSkewedPoints(size, 1, 3)
#     getSkewedPoints(size, 3, 3)
#     getSkewedPoints(size, 5, 3)
#     getSkewedPoints(size, 7, 3)
#     getSkewedPoints(size, 9, 4)
```
